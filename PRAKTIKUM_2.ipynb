```python
class mahasiswa:
  universitas="universitas muhammadiyah semarang"
  total_mahasiswa=0

  def __init__(self,nama: str,nim: str,prodi: str):
    self.nama=nama.strip().title()
    self.nim=nim.strip()
    self.prodi=prodi.strip()
    mahasiswa.total_mahasiswa +=1

  def tampilkan_profil(self):
    print(f"Nama: {self.nama}")
    print(f"Nim : {self.nim}")
    print(f"Prodi : {self.prodi}")
    print(f"univ : {mahasiswa.universitas}")
    print("-" * 40)

  @classmethod
  def jumlah_mahasiswa(cls):
    print(f"Total mahasiswa terdaftar: {cls.total_mahasiswa}")

  def main():
    daftar_mahasiswa: list[mahasiswa] =[]

    menu="""
 === sistem data mahasiswa ===
    1. Tambah mahasiswa
    2. tampilkan data mahasiswa
    3. tampilkan jumlah mahasiswa
    4. ubah nama universitas
    5. keluar
    pilih: """
    while True:
      pilihan=input(menu).strip()
      if pilihan=="1":
        nama=input("nama: ")
        nim=input("nim: ")
        prodi=input("prodi: ")
        m=mahasiswa(nama,nim,prodi)
        daftar_mahasiswa.append(m)
        print("mahasiswa berhasil ditambahkan. \n")
      elif pilihan =="2":
        if not daftar_mahasiswa:
          print("belum ada data. \n")
        else:
          for m in daftar_mahasiswa:
            m.tampilkan_profil()
      elif pilihan == "3":
        mahasiswa.jumlah_mahasiswa()
        print()
      elif pilihan =="4":
        baru=input("masukan nama universitas baru: ").strip()
        if baru:
          mahasiswa.universitas =baru
          print("class atribute 'universitas' berhasil diperbarui. \n")
      elif pilihan=="5":
        print("selesai. terima kasih")
        break
      else:
        print("pilihan tidak valid. \n")

  if __name__=="__main__":
    main()







```

```python

from typing import Literal

class reservasi_syariah:
  hotel_nama="hotel syariah unimus"
  total_reservasi =0
  kapasitas_kamar={"standard":10,"deluxe":6,"family":4}

  def __init__(self,nama_tamu: str,jenis_kelamin: str,pasangan_seruangan: bool, status_menikah: bool,tipe_kamar: str,malam:int):
    self.nama_tamu=nama_tamu.strip().title()
    self.jenis_kelamin=jenis_kelamin.strip().upper()
    self.pasangan_seruangan=pasangan_seruangan
    self.status_menikah=status_menikah
    self.tipe_kamar=tipe_kamar.strip().lower()
    self.malam=malam
    reservasi_syariah.total_reservasi +=1


  def tampilkan_reserfasi(self):
    print(f"Nama Tamu: {self.nama_tamu}")
    print(f"Jenis Kelamin: {self.jenis_kelamin}")
    print(f"Pasangan Seruangan: {self.pasangan_seruangan}")
    print(f"Status Menikah: {self.status_menikah}")
    print(f"Tipe Kamar: {self.tipe_kamar}")
    print(f"Lama Menginap: {self.malam}")
    print(f"Hotel: {reservasi_syariah.hotel_nama}")
    if self.pasangan_seruangan and not self.status_menikah:
      print(f"status : tidak diterima")
    else:
      print(f"status : diterima")
    print("-" * 40)

  @classmethod
  def hitung_total_reservasi(cls):
    print(f"Total reservasi: {cls.total_reservasi}")


def main():
  daftar_reservasi: list[reservasi_syariah] =[]

  menu="""
 === sistem reservasi syariah ===
  1. buat reservasi
  2. lihat semua reservasi
  3. tampilkan total reservasi diterima
  4. ganti nama hotel
  5. info fasilitas syariah
  6. keluar
  pilih: """
  while True:
    choose=input(menu).strip()
    if choose=="1":
      nama_tamu=input("tamu reservasi: ")
      jenis_kelamin=input("jenis kelamin(L / P): ")
      pasangan_seruangan_input=input("pasangan seruangan (y / n): ").strip().lower()
      pasangan_seruangan = pasangan_seruangan_input == 'y'
      status_menikah_input=input("status menikah (y / n): ").strip().lower()
      status_menikah = status_menikah_input == 'y'
      tipe_kamar=input("tipe kamar (standard / deluxe / family): ")
      try:
          malam=int(input("lama menginap: "))
      except ValueError:
          print("Input lama menginap harus berupa angka.")
          continue
      r=reservasi_syariah(nama_tamu, jenis_kelamin, pasangan_seruangan, status_menikah, tipe_kamar, malam)
      daftar_reservasi.append(r)
      print("reservasi berhasil ditambahkan. \n")

    elif choose =="2":
      if not daftar_reservasi:
        print("belum ada data. \n")
      else:
        for r_item in daftar_reservasi:
          r_item.tampilkan_reserfasi()
    elif choose == "3":
      if not daftar_reservasi:
        print("belum ada data. \n")
      else:
        accepted_count = 0
        for r_item in daftar_reservasi:
            if not (r_item.pasangan_seruangan and not r_item.status_menikah):
                accepted_count += 1
        print(f"Total reservasi diterima: {accepted_count}")
        print()
    elif choose =="4":
      baru=input("masukan nama hotel baru: ").strip()
      if baru:
        reservasi_syariah.hotel_nama =baru
        print("class atribute 'hotel' berhasil diperbarui. \n")
    elif choose=="5":
      print(f"info fasilitas hotel : ")
      print(f"1.penanda kiblat ")
      print(f"2. jadwal solat ")

    elif choose=="6":
      print("terima kasih sudah reservasi. ")
      break
    else:
      print("pilihan tidak valid. \n")

if __name__=="__main__":
  main()
```
