```python
class Buku:
    def __init__(self, judul, penulis, tahun):
        self.judul = judul
        self.penulis = penulis
        self.tahun = tahun

    def info(self):
        return f"Judul: {self.judul}, Penulis: {self.penulis}, Tahun: {self.tahun}"


# daftar object buku
daftar_buku = [
    Buku("Algoritma", "Munir", 2010),
    Buku("Machine Learning", "Goodfellow", 2016),
    Buku("Python Dasar", "Zed Shaw", 2015)
]

# tampilkan informasi semua buku
for buku in daftar_buku:
    print(buku.info())
```

```python
class mahasiswa:
  nilai_bobot={
      "A":4.0,"A-":3.7,
      "B+":3.3,"B":3.0,"B-":2.7,
      "C+":2.3,"C":2.0,"C-":1.7,
      "D":1.0,"E":0.0

  }

  def __init__(self,nama,nim,jurusan):
    self.nama=nama
    self.nim=nim
    self.jurusan=jurusan


  def perkenalan(self):
    print(f"halo, Nama saya {self.nama} ({self.nim}) dari {self.jurusan}")


  def update_jurusan(self,new_jurusan):
    self.new_jurusan=new_jurusan

  def hitung_ipk(self,course):
    total_sks=0
    total_bobot=0.0
    for c in course:
      sks=int(c["sks"])
      nilai_huruf=str(c["nilai"]).upper().strip()
      if nilai_huruf not in self.nilai_bobot:
        raise ValueError(f"Nilai {nilai_huruf} tidak valid")
      bobot=self.nilai_bobot[nilai_huruf]
      total_sks+=sks
      total_bobot+=bobot * sks
    if total_sks == 0:
      return 0.0
    return round(total_bobot/total_sks, 2)

nouval =mahasiswa("nouval","1182420050","infor")
ghifary=mahasiswa("ghifary","13182420049","infor")

# Example of how to call hitung_ipk with a list of courses
courses_taken = [
    {"sks": 4, "nilai": "b"},
    {"sks": 2, "nilai": "B-"},
    {"sks": 4, "nilai": "C"}
]
ipk_nouval = nouval.hitung_ipk(courses_taken)
print(f"IPK {nouval.nama}: {ipk_nouval}")

nouval.perkenalan()
ghifary.perkenalan()
```
