```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error

from abc import ABC, abstractmethod
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# Atur random seed untuk reproduksibilitas
np.random.seed(42)
import tensorflow as tf
tf.random.set_seed(42)

# ==============================================================================
# 1. Abstract Class – BasePipeline (PERBAIKAN: Menambahkan **kwargs)
# ==============================================================================

class BasePipeline(ABC):
    """Abstract Class yang mendefinisikan kerangka kerja dasar untuk ML pipeline."""
    # PERBAIKAN 1: Menerima keyword arguments tambahan
    def __init__(self, data_path, look_back=60, train_ratio=0.8, **kwargs):
        self.data_path = data_path
        self.look_back = look_back
        self.train_ratio = train_ratio
        self.raw_data = None
        self.model = None

    @abstractmethod
    def load_data(self): pass

    @abstractmethod
    def preprocess(self): pass

    @abstractmethod
    def build_model(self): pass

    @abstractmethod
    def train(self): pass

    @abstractmethod
    def evaluate(self): pass

# ==============================================================================
# 2. Class StockDataLoader (PERBAIKAN: Menambahkan dan Meneruskan **kwargs)
# ==============================================================================

class StockDataLoader(BasePipeline):
    """Kelas turunan BasePipeline untuk memuat, downsample, dan mempersiapkan data saham."""
    # PERBAIKAN 2: Menerima **kwargs dan meneruskannya ke BasePipeline
    def __init__(self, data_path, look_back=60, train_ratio=0.8, **kwargs):
        # Meneruskan semua argumen ke BasePipeline, termasuk yang tidak dikenali
        super().__init__(data_path, look_back, train_ratio, **kwargs)
        self.scaled_data = None
        self.train_data = None
        self.test_data = None
        self.X_train, self.y_train = None, None
        self.X_test, self.y_test = None, None

    def load_data(self):
        """Membaca CSV, downsample ke frekuensi harian, dan mengambil kolom harga penutupan ('close')."""
        print(f"--- 1. Memuat Data dari: {self.data_path} ---")
        try:
            df = pd.read_csv(self.data_path, index_col='timestamp', parse_dates=True)
            df_daily = df['close'].resample('D').last().dropna()

            self.raw_data = df_daily.values.reshape(-1, 1)
            print(f"Data Berhasil Dimuat dan di-Downsample ke Harian. Total {len(self.raw_data)} baris.")
        except FileNotFoundError:
            print(f"ERROR: File {self.data_path} tidak ditemukan. Pastikan file ada.")
            raise

    def preprocess(self):
        print("--- 2. Pra-pemrosesan Data (Dilanjutkan oleh TimeSeriesPreprocessor) ---")
        pass

    def build_model(self): pass
    def train(self): pass
    def evaluate(self): pass


# ==============================================================================
# 3. Class TimeSeriesPreprocessor
# ==============================================================================

class TimeSeriesPreprocessor:
    """Kelas untuk melakukan scaling dan splitting data time-series."""
    def __init__(self, data, train_ratio=0.8):
        self.data = data
        self.train_ratio = train_ratio
        self._scaler = MinMaxScaler(feature_range=(0, 1)) # Protected attribute
        self.scaled_data = None
        self.train_data = None
        self.test_data = None

    def scale_data(self):
        self.scaled_data = self._scaler.fit_transform(self.data)
        return self.scaled_data

    def split_data(self):
        train_size = int(len(self.scaled_data) * self.train_ratio)
        self.train_data = self.scaled_data[:train_size]
        self.test_data = self.scaled_data[train_size:]
        print(f"Ukuran Training: {len(self.train_data)}, Ukuran Testing: {len(self.test_data)}")
        return self.train_data, self.test_data

    def inverse_transform(self, scaled_array):
        return self._scaler.inverse_transform(scaled_array)

# ==============================================================================
# 4. Class SequenceBuilder
# ==============================================================================

class SequenceBuilder:
    """Kelas untuk mengubah data time-series (scaled) menjadi format supervised learning (X dan y) untuk LSTM."""
    def __init__(self, data, look_back):
        self.data = data
        self.look_back = look_back

    def create_sequences(self):
        X, y = [], []
        for i in range(self.look_back, len(self.data)):
            X.append(self.data[i - self.look_back:i, 0])
            y.append(self.data[i, 0])

        X = np.array(X)
        y = np.array(y)
        X = np.reshape(X, (X.shape[0], X.shape[1], 1))

        return X, y

# ==============================================================================
# 5. Class LSTMModelFactory
# ==============================================================================

class LSTMModelFactory:
    """Kelas untuk membangun dan mengompilasi arsitektur model LSTM."""
    def __init__(self, timesteps, features, lstm_units=[50, 50], dense_units=25):
        self.timesteps = timesteps
        self.features = features
        self.lstm_units = lstm_units
        self.dense_units = dense_units

    def build_model(self):
        """Membangun arsitektur LSTM minimal: 2 LSTM, 1 Dense hidden, 1 Dense output."""
        print("--- 3. Membangun Model LSTM ---")
        model = Sequential()
        model.add(LSTM(self.lstm_units[0], return_sequences=True,
                       input_shape=(self.timesteps, self.features)))
        model.add(LSTM(self.lstm_units[1], return_sequences=False))
        model.add(Dense(self.dense_units))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
        print("Model Berhasil Dikompilasi. Arsitektur:")
        model.summary()
        return model

# ==============================================================================
# 6. Class Trainer
# ==============================================================================

class Trainer:
    """Kelas untuk melatih model dengan custom callback Early Stopping."""
    def __init__(self, model, X_train, y_train, epochs=100, batch_size=32, validation_split=0.1):
        self.model = model
        self.X_train = X_train
        self.y_train = y_train
        self.epochs = epochs
        self.batch_size = batch_size
        self.validation_split = validation_split
        self.history = None

    def train_model(self):
        """Melatih model dan menggunakan Early Stopping."""
        print("--- 4. Melatih Model ---")
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        self.history = self.model.fit(
            self.X_train, self.y_train,
            epochs=self.epochs,
            batch_size=self.batch_size,
            validation_split=self.validation_split,
            callbacks=[early_stopping],
            verbose=1
        )
        print("Pelatihan Selesai.")
        return self.history

# ==============================================================================
# 7. Class Evaluator
# ==============================================================================

class Evaluator:
    """Kelas untuk melakukan prediksi, menghitung metrik, dan menampilkan grafik."""
    def __init__(self, model, preprocessor, X_train, y_train, X_test, y_test, look_back):
        self.model = model
        self.preprocessor = preprocessor
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test
        self.look_back = look_back

    def predict(self, X):
        """Melakukan prediksi pada data input."""
        return self.model.predict(X, verbose=0)

    def evaluate_performance(self):
        """Menghitung MAE untuk data training dan testing."""
        # Prediksi
        train_predict_scaled = self.predict(self.X_train)
        test_predict_scaled = self.predict(self.X_test)

        # Inverse Transform
        train_actual = self.preprocessor.inverse_transform(self.y_train.reshape(-1, 1))
        test_actual = self.preprocessor.inverse_transform(self.y_test.reshape(-1, 1))
        train_predict = self.preprocessor.inverse_transform(train_predict_scaled)
        test_predict = self.preprocessor.inverse_transform(test_predict_scaled)

        # Hitung MAE
        train_mae = mean_absolute_error(train_actual, train_predict)
        test_mae = mean_absolute_error(test_actual, test_predict)

        print("--- 5. Evaluasi Model ---")
        print(f"MAE Training (Original Scale): {train_mae:.4f}")
        print(f"MAE Testing (Original Scale): {test_mae:.4f}")

        self.plot_results(train_predict_scaled, test_predict_scaled,
                          train_actual, test_actual, train_predict, test_predict)

    def plot_results(self, train_pred_scaled, test_pred_scaled,
                     train_actual, test_actual, train_pred, test_pred):
        """Menampilkan dua grafik: scaled dan original scale."""

        full_data = self.preprocessor.data
        train_size = len(train_actual)
        test_size = len(test_actual)

        # Plot 1: Aktual vs Prediksi (Scaled)
        plt.figure(figsize=(12, 6))
        train_plot = np.empty_like(full_data)
        train_plot[:, :] = np.nan
        train_plot[self.look_back:self.look_back + train_size, :] = train_pred_scaled

        test_plot = np.empty_like(full_data)
        test_plot[:, :] = np.nan
        test_plot[train_size + self.look_back:train_size + self.look_back + test_size, :] = test_pred_scaled

        plt.plot(self.preprocessor.scaled_data, label='Actual Data (Scaled)')
        plt.plot(train_plot, label='Train Prediction (Scaled)')
        plt.plot(test_plot, label='Test Prediction (Scaled)')
        plt.title('Actual vs Prediction (Scaled Data)')
        plt.xlabel('Time Step (Daily)')
        plt.ylabel('Price (Scaled)')
        plt.legend()
        plt.savefig('actual_vs_prediction_scaled.png')

        # Plot 2: Aktual vs Prediksi (Original Scale)
        plt.figure(figsize=(12, 6))
        actual_prices = full_data

        train_plot_orig = np.empty_like(actual_prices)
        train_plot_orig[:, :] = np.nan
        train_plot_orig[self.look_back:self.look_back + train_size, :] = train_pred

        test_plot_orig = np.empty_like(actual_prices)
        test_plot_orig[:, :] = np.nan
        test_plot_orig[train_size + self.look_back:train_size + self.look_back + test_size, :] = test_pred

        plt.plot(actual_prices, label='Actual Price (Original)')
        plt.plot(train_plot_orig, label='Train Prediction (Original)')
        plt.plot(test_plot_orig, label='Test Prediction (Original)')
        plt.title('Actual vs Prediction (Original Scale)')
        plt.xlabel('Time Step (Daily)')
        plt.ylabel('Price (IDR)')
        plt.legend()
        plt.savefig('actual_vs_prediction_original.png')


# ==============================================================================
# Main Execution Class (PERBAIKAN: Memanggil super().__init__ dengan semua kwargs)
# ==============================================================================

class LSTMPredictionPipeline(StockDataLoader):
    """Kelas utama yang mengintegrasikan seluruh komponen pipeline."""
    def __init__(self, data_path, look_back, train_ratio, lstm_units, dense_units):
        # PERBAIKAN 3: Meneruskan semua argumen termasuk lstm_units dan dense_units
        super().__init__(data_path, look_back, train_ratio, lstm_units=lstm_units, dense_units=dense_units)
        self.lstm_units = lstm_units
        self.dense_units = dense_units
        self.preprocessor = None
        self.trainer = None
        self.evaluator = None

    def preprocess(self):
        # TimeSeriesPreprocessor
        self.preprocessor = TimeSeriesPreprocessor(self.raw_data, self.train_ratio)
        scaled_data = self.preprocessor.scale_data()
        train_scaled, test_scaled = self.preprocessor.split_data()

        # SequenceBuilder (Training Data)
        seq_builder_train = SequenceBuilder(train_scaled, self.look_back)
        self.X_train, self.y_train = seq_builder_train.create_sequences()

        # SequenceBuilder (Testing Data)
        total_train_size = len(train_scaled)
        all_test_data = scaled_data[total_train_size - self.look_back:]
        seq_builder_test = SequenceBuilder(all_test_data, self.look_back)
        self.X_test, self.y_test = seq_builder_test.create_sequences()

        print(f"Shape X_train: {self.X_train.shape}, y_train: {self.y_train.shape}")
        print(f"Shape X_test: {self.X_test.shape}, y_test: {self.y_test.shape}")

    def build_model(self):
        # LSTMModelFactory
        timesteps = self.X_train.shape[1]
        features = self.X_train.shape[2]
        model_factory = LSTMModelFactory(timesteps, features, self.lstm_units, self.dense_units)
        self.model = model_factory.build_model()

    def train(self):
        # Trainer
        self.trainer = Trainer(self.model, self.X_train, self.y_train, epochs=100, batch_size=32)
        self.history = self.trainer.train_model()

    def evaluate(self):
        # Evaluator
        self.evaluator = Evaluator(self.model, self.preprocessor, self.X_train, self.y_train,
                                   self.X_test, self.y_test, self.look_back)
        self.evaluator.evaluate_performance()

    def run_pipeline(self):
        self.load_data()
        self.preprocess()
        self.build_model()
        self.train()
        self.evaluate()


# ==============================================================================
# EKSEKUSI
# ==============================================================================

# Ganti nama file menjadi 'ANTM.csv'
DATA_FILE = 'ANTM.csv'

# Eksperimen 1: Base Model (look_back = 60)
LOOK_BACK_EXP1 = 60
LSTM_UNITS_EXP1 = [50, 50]
TRAIN_RATIO_EXP1 = 0.8

print("=== Eksperimen 1: Base Model (look_back=60) ===")
try:
    pipeline_exp1 = LSTMPredictionPipeline(
        data_path=DATA_FILE,
        look_back=LOOK_BACK_EXP1,
        train_ratio=TRAIN_RATIO_EXP1,
        lstm_units=LSTM_UNITS_EXP1,
        dense_units=25
    )
    pipeline_exp1.run_pipeline()

    # Eksperimen 2: Mengubah Look_Back
    LOOK_BACK_EXP2 = 30

    print("\n\n=== Eksperimen 2: Mengubah Look_Back (look_back=30) ===")
    pipeline_exp2 = LSTMPredictionPipeline(
        data_path=DATA_FILE,
        look_back=LOOK_BACK_EXP2,
        train_ratio=TRAIN_RATIO_EXP1,
        lstm_units=LSTM_UNITS_EXP1,
        dense_units=25
    )
    pipeline_exp2.run_pipeline()

except Exception as e:
    print(f"\n[ERROR saat menjalankan Pipeline]: {e}")
```
