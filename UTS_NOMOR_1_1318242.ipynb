```python
#UTS NOMOR 1
class mahasiswa:
  def __init__(self,nim,nama,prodi,nilai):
    self.nama = nama
    self.nim = nim
    self.prodi = prodi
    self.nilai = nilai

  def tampilkan(self):
    print("SAYA ADALAH....")
    print("Nama Mahasiswa : ",self.nama)
    print("NIM : ",self.nim)
    print("Program Studi : ",self.prodi)
    print("Nilai : ",self.nilai)

  def menghitung_rata_rata(self):
    print("Nilai Rata-Rata : ",sum(self.nilai)/len(self.nilai))

  def status_kelulusan(self):
    if sum(self.nilai)/len(self.nilai) >= 75:
      print("Status Kelulusan : LULUS")
    else:
      print("Status Kelulusan : TIDAK LULUS")

#membuat class turunan mahasiswa aktif
class mahasiswa_aktif(mahasiswa):
  def __init__(self, nim, nama, prodi, nilai, semester):
      # Memanggil konstruktor superclass
      super().__init__(nim, nama, prodi, nilai)
      self.semester = semester

  def menampilkan_data(self):
      # Memanggil method asli dari superclass
      super().tampilkan()

      # Menambahkan informasi tambahan
      print(f"Status : Mahasiswa Aktif")
      print(f"Semester: {self.semester}")

class mahasiswaalumni(mahasiswa):
  def __init__(self, nim, nama, prodi, nilai, tahun_lulus):
      super().__init__(nim, nama, prodi, nilai)
      if(tahun_lulus) < 1999:
        raise ValueError("Tahun lulus tidak valid.")
      self.tahun_lulus = tahun_lulus

  def menampilkan_data(self):
      # Memanggil method asli dari superclass
      super().tampilkan()

      # Menambahkan informasi tambahan
      print(f"Status : Mahasiswa Alumni")
      print(f"Tahun Kelulusan: {self.tahun_lulus}")

if __name__ == "__main__":

    print("=== Contoh Mahasiswa Aktif (Lulus) ===")
    try:
        mhs_aktif = mahasiswa_aktif(
            nim="13182420050",
            nama="Nouval Al Ghifary",
            prodi="Teknik Informatika",
            nilai=[80, 90, 95, 85],
            semester=3,

        )
        mhs_aktif.menampilkan_data()
        mhs_aktif.menghitung_rata_rata()
        mhs_aktif.status_kelulusan()

    except (ValueError, TypeError) as e:
        print(f"Input Gagal: {e}")

    print("\n" + "="*40 + "\n")

    print("=== Contoh Mahasiswa Alumni (tidak lulus) ===")
    try:
        mhs_alumni = mahasiswaalumni(
            nim="54321",
            nama="bae suzy",
            prodi="kedokteran",
            nilai=[70, 70, 70, 85],
            tahun_lulus=2023
        )
        mhs_alumni.menampilkan_data()
        mhs_alumni.menghitung_rata_rata()
        mhs_alumni.status_kelulusan()

    except (ValueError, TypeError) as e:
        print(f"Input Gagal: {e}")

    print("\n" + "="*40 + "\n")

    print("=== Contoh Input Gagal (Exception Handling) ===")
    try:
        # Input nilai salah (mengandung string)
        mhs_gagal = mahasiswa(
            nim="99999",
            nama="Gagal Input",
            prodi="Sistem Informasi",
            nilai=[80, "sembilan puluh", 75]
        )
        mhs_gagal.tampilkan() # Base class uses 'tampilkan' not 'menampilkan_data'
        mhs_gagal.menghitung_rata_rata()
        mhs_gagal.status_kelulusan()

    except (ValueError, TypeError) as e:
        print(f"ERROR: Terjadi kesalahan saat membuat objek mahasiswa.")
        print(f"Detail Error: {e}")
```
