```python
class animal:
  def sound(self):
    print("some generic sound")

class dog(animal):
  def sound(self):
    print("woof!")

class cat(animal):
  def sound(self):
    print("meow!")

for obj in[dog(),cat(),animal()]:
  obj.sound()

```

```python
class payment:
  def process_payment(self,amount):
    pass

class banktransfer(payment):
  def process_payment(self,amount):
    print(f"processing bank transfer of {amount}")

class ewallet(payment):
  def process_payment(self,amount):
    print(f"processing e-wallet payment of {amount}")

class creditcard(payment):
  def process_payment(self,amount):
    print(f"processing credit card payment of {amount}")

payment_methods=[banktransfer(),ewallet(),creditcard()]
for method in payment_methods:
  method.process_payment(150000)
```

```python
class hewan:
  def __init__(self,nama):
    self.nama=nama
  def suara(self):
    print("some generic sound")

class kucing(hewan):
  def suara(self):
    return "meow!"

class anjing(hewan):
  def suara(self):
    return"woof!"

class sapi(hewan):
  def suara(self):
    return "mooo"

hewan_list=[
    anjing("buddy"),
    kucing("whiskers"),
    sapi("bessie")
]

for hewan in hewan_list:
  print(hewan.suara())
```

```python
class mobil:
  def start(self):
    return "mesin mobill menyala"

class motor:
  def start(self):
    return "mesin motor menyala"

class perahu:
  def start(self):
    return "mesin perahu menyala"

def operasi_mesin(kendaraan):
  print(kendaraan.start())


kendaraan_list=[
    mobil(),
    motor(),
    perahu()
]

for kendaraan in kendaraan_list:
  operasi_mesin(kendaraan)

```

```python
class kendaraan:
  def start(self):
    return "mesin kendaraan menyala"

  def move(self):
    return "kendaraan mulai bergerak"

  def stop(self):
    return " kendaraan berhenti"

#subclass mobil
class mobil(kendaraan):
  def start(self):
    return "mobil dihidupkan dengan tombol start"

  def move(self):
    return "mobil mulai bergerak"

  def stop(self):
    return "mobil berhenti "

#subclass motor
class motor(kendaraan):
  def start(self):
    return "motor dihidupkan dengan tombol start"

  def move(self):
    return "motor mulai bergerak"

  def stop(self):
    return "motor berhenti "

#subclass kapal
class kapal(kendaraan):
  def start(self):
    return "kapal dihidupkan dengan tombol start"

  def move(self):
    return "kapal mulai bergerak"

  def stop(self):
    return "kapal berhenti "

def operasikan_kendaraan(kendaraan):
  print(kendaraan.start())
  print(kendaraan.move())
  print(kendaraan.stop())

class pesawat:
  def start(self):
    return "pesawat dihidupkan dengan tombol start"

  def move(self):
    return "pesawat mulai bergerak"

  def stop(self):
    return "pesawat berhenti "

kendaraan_list=[
    mobil(),
    motor(),
    kapal(),
    pesawat()
]

for kendaraan in kendaraan_list:
  operasikan_kendaraan(kendaraan)




```
