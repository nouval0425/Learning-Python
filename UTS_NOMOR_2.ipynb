```python
#NOMOR 2 Sistem Kasir Mini

class produk:
  def __init__(self, nama, harga, stok):
    self.nama = nama
    self.__harga = 0 # Initialize, then use setter for validation
    self.set_harga(harga) # Use setter for initial price
    self.stok = stok

  def get_harga(self):
    return self.__harga

  def set_harga(self, harga_baru):
    if harga_baru > 0:
      self.__harga = harga_baru
    else:
      print("ERROR: Harga tidak valid. Harga harus lebih dari 0.")

  def kurangi_stok(self, jumlah):
    if self.stok >= jumlah:
      self.stok -= jumlah
      return True
    else:
      print(f"Stok {self.nama} tidak cukup. Tersedia: {self.stok}, Diminta: {jumlah}.")
      return False

  def tambah_stok(self, jumlah):
    if jumlah > 0:
      self.stok += jumlah
      return True
    else:
      print("Jumlah penambahan stok harus lebih dari 0.")
      return False


class transaksi:
  def __init__(self):
    self.item = {} # Dictionary to store {produk_obj: quantity}
    self.total_belanja = 0

  def tambah_produk(self, produk_obj, jumlah):
    """Menambahkan produk ke keranjang."""
    if not isinstance(produk_obj, produk):
        print("Gagal: Objek yang ditambahkan harus dari kelas produk.")
        return False
    if jumlah <= 0:
        print("Gagal: Jumlah produk harus lebih dari 0.")
        return False

    if produk_obj.kurangi_stok(jumlah):
        # Jika produk sudah ada di keranjang, tambahkan jumlahnya
        self.item[produk_obj] = self.item.get(produk_obj, 0) + jumlah
        print(f"Berhasil: {jumlah}x {produk_obj.nama} ditambahkan ke keranjang.")
        return True
    else:
        print(f"Gagal: {produk_obj.nama} tidak ditambahkan.")
        return False

  def hitung_total(self):
    total = 0
    for produk_obj, jumlah in self.item.items():
        # Menggunakan getter untuk mengakses harga yang dienkapsulasi
        total += produk_obj.get_harga() * jumlah

    self.total_belanja = total
    return total

  def tampilkan_struk(self):
    """Menampilkan struk belanja."""
    print("\n--- Struk Belanja (Non-Member) ---")
    if not self.item:
        print("Keranjang kosong.")
        return

    for produk_obj, jumlah in self.item.items():
        subtotal = produk_obj.get_harga() * jumlah
        print(f"* {produk_obj.nama} (x{jumlah}) - @{produk_obj.get_harga():,} = Rp {subtotal:,}")

    print("-" * 30)
    # Memanggil method hitung_total() versi class ini
    self.hitung_total()
    print(f"TOTAL BAYAR : Rp {self.total_belanja:,.0f}")


# --- 3. Class MemberTransaksi (Subclass dengan Overriding) ---

class MemberTransaksi(transaksi):

  def __init__(self, nama_member, diskon_persen=10):
    # Memanggil __init__ dari superclass (Transaksi)
    super().__init__() # Parent's init now correctly initializes self.item and self.total_belanja
    self.nama_member = nama_member
    self.diskon_persen = diskon_persen
    self.total_sebelum_diskon = 0

  # --- Method Overriding ---

  def hitung_total(self):

    total_awal = super().hitung_total()
    self.total_sebelum_diskon = total_awal


    potongan_diskon = total_awal * (self.diskon_persen / 100)
    total_akhir = total_awal - potongan_diskon

    self.total_belanja = total_akhir
    return total_akhir



  def tampilkan_struk(self):
    """
    Method Overriding:
    Meng-override struk agar menampilkan rincian diskon.
    """
    print(f"\n--- Struk Belanja (Member: {self.nama_member}) ---")
    if not self.item:
        print("Keranjang kosong.")
        return

    for produk_obj, jumlah in self.item.items():
        subtotal = produk_obj.get_harga() * jumlah
        print(f"* {produk_obj.nama} (x{jumlah}) - @{produk_obj.get_harga():,} = Rp {subtotal:,}")

    print("-" * 30)

    # Memanggil method hitung_total() versi MemberTransaksi
    self.hitung_total()

    print(f"Subtotal    : Rp {self.total_sebelum_diskon:,.0f}")
    print(f"Diskon ({self.diskon_persen}%) : -Rp {self.total_sebelum_diskon * (self.diskon_persen / 100):,.0f}")
    print(f"TOTAL BAYAR : Rp {self.total_belanja:,.0f}")


# --- 4. Contoh Penggunaan ---

if __name__ == "__main__":
    # 1. Buat beberapa produk
    motor = produk(nama="Motor", harga=500000, stok=50)
    pensil = produk(nama="Pensil 2B", harga=2000, stok=100)
    penghapus = produk(nama="Penghapus", harga=1000, stok=30)
    monitor = produk(nama="Monitor Gaming", harga=2500000, stok=5)

    # 2. Skenario 1: Transaksi Non-Member
    print("=== Skenario 1: Transaksi Non-Member ===")
    transaksi_biasa = transaksi()
    transaksi_biasa.tambah_produk(motor, 3)
    transaksi_biasa.tambah_produk(pensil, 5)

    # Coba tambahkan produk melebihi stok
    transaksi_biasa.tambah_produk(penghapus, 50) # This should fail

    # Tampilkan struk
    transaksi_biasa.tampilkan_struk()  # Total should be 20000

    print("\n" + "="*40 + "\n")

    # 3. Skenario 2: Transaksi Member (Method Overriding terjadi di sini)
    print("=== Skenario 2: Transaksi Member (Diskon 10%) ===")
    transaksi_member = MemberTransaksi(nama_member="Andi", diskon_persen=10)
    transaksi_member.tambah_produk(motor, 4)     # 4 x 5000 = 20000
    transaksi_member.tambah_produk(penghapus, 5) # 5 x 1000 = 5000
    transaksi_member.tambah_produk(monitor, 1) # 1 x 2500000 = 2500000
                                                # Subtotal = 25000 + 2500000 = 2525000
                                                # Diskon 10% = -252500
                                                # Total = 2272500
    # Tampilkan struk
    transaksi_member.tampilkan_struk()

    print("\n" + "="*40 + "\n")

    # 4. Bukti Encapsulation
    print("=== Bukti Encapsulation ===")
    try:
        # Mencoba mengakses atribut 'private' secara langsung (akan gagal)
        print(f"Harga motor (akses langsung): {motor.__harga}")
    except AttributeError as e:
        print(f"Error akses langsung: {e}")

    # Cara yang benar (menggunakan getter)
    print(f"Harga motor (getter): Rp {motor.get_harga():,}")

    # Test set_harga with invalid value
    print("\n=== Testing set_harga dengan nilai tidak valid ===")
    motor.set_harga(-100) # Should print error
    print(f"Harga motor setelah mencoba set -100: Rp {motor.get_harga():,}") # Should remain 5000
    motor.set_harga(5500) # Should set new price
    print(f"Harga motor setelah set 5500: Rp {motor.get_harga():,}")
```
