```python
#encapsulasi digunakan untuk menutup code agar code lebih aman menggunakan class

class mahasiswa:
  def __init__(self,name,nim):
    self.__name = name
    self.__nim = nim

#setter = umtuk mengubah dari luar
  def set_name(self,name):
    self.__name = name

  def set_nim(self,nim):
    self.__nim = nim

#getter = untuk mengambil nilai
  def get_name(self):
    return self.__name

  def get_nim(self):
    return self.__nim

mhs1 = mahasiswa("ade","1318242000")

print("Nama :",mhs1.get_name())
print("NIM :",mhs1.get_nim())

mhs1.set_name("nizar");
mhs1.set_nim("1318242001")

print("Nama :",mhs1.get_name())
print("NIM :",mhs1.get_nim())



```

```python
class person:
  def __init__(self,nama,age):
  #variabel class tidak dapat diaccses diluar class karena private dan harus menggunakan getter dan setter
    self.__nama = nama
    self.__age = age

  def set_nama(self,nama):
    self.__nama = nama

  def set_age(self,age):
    self.__age = age

  def get_nama(self):
    return self.__nama

  def get_age(self):
    return self.__age

person1 = person("annas","50")
print("nama :",person1.get_nama())
print("umur :",person1.get_age())

print("\nsetelah diubah----")
person1.set_nama("khoirul")
print("nama :",person1.get_nama())
print("umur :",person1.get_age())
```

```python
class vehicle:
  def __init__(self,merk,kecepatan):
    self.__merk = merk
    self.__kecepatan = kecepatan

  def set_merk(self,merk_baru):
    self.__merk = merk_baru

  def set_kecepatan(self,kecepatan_baru):
    if kecepatan_baru > 0:
      self.__kecepatan = kecepatan_baru
    else:
      print("kecepatan tidak boleh negatif")

  def get_merk(self):
    return self.__merk

  def get_kecepatan(self):
    return self.__kecepatan

mobil = vehicle("toyota",200)
print("merk :",mobil.get_merk())
print("kecepatan :",mobil.get_kecepatan())
```

```python
class akunbank:
  def __init__(self,nama,saldo):
    self.__nama = nama
    self.__saldo = saldo

#getter
  def get_nama(self):
    return self.__nama

  def get_saldo(self):
    return self.__saldo

#setter
  def set_nama(self,nama_baru):
    self.__nama = nama_baru

  def set_saldo(self,saldo_baru):
    self.__saldo = saldo_baru

#setter untuk menarik saldo
  def tarik(self,jumlah):
    if 0 < jumlah <= self.__saldo:
      self.__saldo -= jumlah
    else:
      print("saldo tidak cukup")

#setter untuk menambah saldo
  def setor(self,jumlah):
    if jumlah > 0:
      self.__saldo += jumlah
    else:
      print("jumlah setor harus lebih dari 0")

  def tampilkan_info(self):
    print("nama : ",self.__nama)
    print("saldo : ",self.__saldo)

  #==========================
  #program utama

print("=== sistem akun bank===")
nama = input("masukan nama : ")
saldo = int(input("masukan saldo : "))

#membuat objek akun
akun1 = akunbank(nama,saldo)

while True:
  print("\n==menu==")
  print("1.tampilkan saldo")
  print("2.tarik saldo")
  print("3.setor saldo")
  print("4.keluar")

  pilihan = input("pilih menu : ")

  if pilihan == "1":
    akun1.tampilkan_info()
  elif pilihan == "2":
    jumlah = int(input("masukan jumlah saldo yang akan ditarik : "))
    akun1.tarik(jumlah)
  elif pilihan == "3":
    jumlah = int(input("masukan jumlah saldo yang akan disetor : "))
    akun1.setor(jumlah)
  elif pilihan == "4":
    print("terimakasih telah menggunakan sistem kami")
    break
  else:
    print("pilihan tidak valid")




```

```python
#TUGAS........
class mobil:
  def __init__(self,plat_nomor,merk,harga_sewa_per_hari,status):
    self.__plat_nomor = plat_nomor
    self.__merk = merk
    self.__harga_sewa_per_hari = harga_sewa_per_hari
    self.__status = status

  def get_plat_nomor(self):
    return self.__plat_nomor

  def get_merk(self):
    return self.__merk

  def get_harga_sewa(self):
    return self.__harga_sewa_per_hari

  def get_status(self):
    return self.__status

  def set_status(self,status_baru):
    if status_baru == "tersedia" or status_baru == "disewa":
      self.__status = status_baru
    else:
      print("status tidak valid")

  def tampilkan_info(self):
    print("plat nomor : ",self.__plat_nomor)
    print("merk : ",self.__merk)
    print("harga sewa per hari : ",self.__harga_sewa_per_hari)
    print("status : ",self.__status)
    print("\n")

class penyewaan:
  def __init__(self,nama_penyewa,mobil,lama_sewa,total_biaya):
    self.__nama_penyewa = nama_penyewa
    self.__mobil = mobil
    self.__lama_sewa = lama_sewa
    self.__total_biaya = total_biaya

  def hitung_total_biaya(self):
    self.__total_biaya = self.__mobil.get_harga_sewa() * self.__lama_sewa

  def tampilkan_info(self):
    print("nama penyewa : ",self.__nama_penyewa)
    print("mobil : ",self.__mobil.get_merk())
    print("lama sewa : ",self.__lama_sewa)
    print("total biaya : ",self.__total_biaya)

  def get_mobil(self): # Add the getter for the mobil object
      return self.__mobil

#=============================
#program utama

#membuat daftar mobil
daftar_mobil = []
daftar_mobil.append(mobil("abc123","honda",50000,"tersedia"))
daftar_mobil.append(mobil("def456","toyota",60000,"tersedia"))
daftar_mobil.append(mobil("ghi789","suzuki",70000,"tersedia"))
daftar_mobil.append(mobil("jkl123","ferari",300000,"tersedia"))
daftar_mobil.append(mobil("lmn456","porsche",100000,"tersedia"))
daftar_mobil.append(mobil("opq789","gtR-86",79000,"tersedia"))

daftar_penyewaan = []

while True:
  print("\n=== sistem penyewaan mobil ===")
  print("1.menampilkan daftar mobil")
  print("2.menyewa mobbil")
  print("3.mengembalikan mobil")
  print("4.menampilkan transaksi penyewaan")
  print("5.keluar")
  pilihan = input("pilih menu : ")

  if pilihan == "1":
    print("=====daftar mobil yang tersedia====")
    for mobil in daftar_mobil:
      mobil.tampilkan_info()

  elif pilihan == "2":
    nama_penyewa = input("masukan nama penyewa : ")
    while True:
      try:
        lama_sewa = int(input("masukan lama sewa : "))
        break
      except ValueError:
        print("Input tidak valid. Masukan angka untuk lama sewa.")

    plat_nomor_sewa = input("masukan plat nomor mobil yang ingin disewa: ")
    mobil_disewa = None
    for mobil in daftar_mobil:
        if mobil.get_plat_nomor() == plat_nomor_sewa and mobil.get_status() == "tersedia":
            mobil_disewa = mobil
            break

    if mobil_disewa:
        mobil_disewa.set_status("disewa")
        penyewaan_baru = penyewaan(nama_penyewa, mobil_disewa, lama_sewa, 0)
        penyewaan_baru.hitung_total_biaya()
        daftar_penyewaan.append(penyewaan_baru)
        print("Mobil berhasil disewa!")
    else:
        print("Mobil tidak ditemukan atau tidak tersedia untuk disewa.")


  elif pilihan == "3":
    plat_nomor_kembali = input("masukan plat nomor mobil yang ingin dikembalikan : ")
    mobil_kembali = None
    for penyewaan in daftar_penyewaan:
        if penyewaan.get_mobil().get_plat_nomor() == plat_nomor_kembali:
            mobil_kembali = penyewaan.get_mobil()
            break

    if mobil_kembali:
        mobil_kembali.set_status("tersedia")
        for penyewaan in daftar_penyewaan:
            if penyewaan.get_mobil() == mobil_kembali:
                daftar_penyewaan.remove(penyewaan)
                print("Mobil berhasil dikembalikan!")
                break

  elif pilihan == "4":
    print("=====daftar transaksi penyewaan====")
    if daftar_penyewaan:
        for sewa in daftar_penyewaan:
            sewa.tampilkan_info()
            print("-" * 20)
    else:
        print("Belum ada transaksi penyewaan.")


  elif pilihan == "5":
    print("terimakasih masih mempercayai kami dalam proses peminjaman mobil")
    break

  else:
    print("pilihan yang kamu pilih tidak valid")
```
